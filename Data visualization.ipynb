#### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
iris_dataset = datasets.load_iris()
```

```python
X= iris_dataset.data
Y = iris_dataset.target
```

```python
sc = StandardScaler()
scalar_transform= sc.fit_transform(X)

pca = decomposition.PCA(n_components = 3).fit_transform(scalar_transform)

fig = plt.figure(2, figsize=(8, 8))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=20, azim=130)
plt.cla()

ax.scatter(pca[:, 0], pca[:, 1], pca[:, 2], c=Y, cmap = plt.cm.get_cmap("rainbow"), edgecolor='k')
plt.show()
```
